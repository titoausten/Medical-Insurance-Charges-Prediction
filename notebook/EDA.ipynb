### Import required libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read Dataset

```python
data = pd.read_csv("insurance.csv")
data.head()
```

<hr>

### Relationship between “smoker” and “charges”

```python
plt.figure(figsize=(6,6))
sns.barplot(x='smoker', y='charges', data=data)
plt.title('the relationship between “smoker” and “charges”')
```

##### Those who smoke pay more charges.

<hr>

### Relationship between “smoker” and “region”

```python
sns.set_style("whitegrid")
sns.catplot('smoker', col='region', data=data, kind='count', aspect=0.6)
```

##### From the Chart, southeast has the highest number of smokers.

<hr>

### Relationship between “sex” and “bmi”

```python
plt.figure(figsize=(6,6))
sns.barplot(x='sex', y='bmi', data=data)
plt.title('the relationship between “sex” and “bmi”')
```

##### Male gender weigh slightly more than the females

<hr>

### Relationship between “age” and “bmi”

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='age', y='bmi', data=data)
plt.title('the relationship between “age” and “bmi”')
```

##### From the scatterplot, it can be inferred that those below 40 tend to weigh more than those above 40.

<hr>

### Relationship between “bmi” and “children”

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='bmi', y='children', data=data)
plt.title('the relationship between “bmi” and “children”')
```

<hr>

### Relationship between “bmi” and “charges”

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='bmi', y='charges', data=data)
plt.title('the relationship between “bmi” and “charges”')
```

<hr>

### Relationship between “region” and “children”

```python
plt.figure(figsize=(6,6))
sns.barplot(x='region', y='children', data=data)
plt.title('the relationship between “region” and “children”')
```

##### southwest and northwest have the most number of children and the difference is very little.

<hr>

### Relationship between region, smoker and bmi using bar plot

```python
sns.set_style("whitegrid")
sns.catplot(x='smoker', y='bmi', data=data, kind="bar", hue='region', aspect=1.55, ci=None,
            hue_order=['southwest', 'southeast', 'northwest', 'northeast'])
plt.title('Relationship between region, smoker and bmi')
```

<hr>

### Checking for an outlier in the "bmi" variable

```python
sns.boxplot(x='bmi', data=data)
```

##### There are outliers as seen from the box plot but they are important and useful for this data

<hr>

### Label encode categorical features

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()

#Categorical Features: sex, smoker and region
data['sex'] = le.fit_transform(data['sex'])
data['smoker'] = le.fit_transform(data['smoker'])

data.head()
```

<hr>

### Correlation between the features and output

```python
plt.figure(figsize=(16,10))
sns.heatmap(data.corr(), annot=True)
plt.title("Heatmap showing the correlation between the features")
plt.show()
```

##### From the Heatmap, it can be seen that smoker feature has the most impact on charges, followed by age and bmi.
